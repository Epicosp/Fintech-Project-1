```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.style as style
```

```python
style.use('fivethirtyeight')
```

```python
df = pd.read_csv("f1db_csv/Formula1_Budgets.csv")
```

```python
df.append(string)
df
```

```python
for col in df.columns[1:]:
    df[col] = df[col].str.replace('$', '')
    df[col] = df[col].astype('float')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.set_index('Year',inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(15,10))
for col in df.columns:
    plt.plot(df[col], label=col)
plt.legend()
plt.title("Formula-1 Budget Spent from 2017 to 2021")
```

```python
df.head(2)
```

```python
df.reset_index(inplace=True)
```

```python
df_f = pd.melt(df, id_vars='Year')
```

```python
df_f
```

```python
op = df_f.groupby("Year").agg(average_budget_all =('value','mean'))
op
```

```python
plt.figure(figsize=(15,10))
sns.barplot(op.index, op["average_budget_all"])
plt.title("Average budget spent by all the companies")
plt.ylabel("Average Budget spent in $")

```

```python
df_f.head(2)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Year', y='value', hue='variable', data=df_f)
plt.title("Formula-1 Budget Spent from 2017 to 2021")
plt.xlabel("Year")
plt.ylabel("Budget Spent in $")
```
